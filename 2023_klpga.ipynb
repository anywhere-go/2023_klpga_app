```python
df_prophet = df_klpga.rename(columns= {'Birth':'ds','Total Win':'y'})
```

```python
from prophet import Prophet
```

```python
m = Prophet()
```

```python
m.fit(df_prophet) #학습 fit
```

```python
m.make_future_dataframe(periods = 365, freq= 'D')
```

```python
future = m.make_future_dataframe(periods = 365, freq= 'D')
```

```python
forecast = m.predict(future)
```

```python
forecast
```

```python
m.plot(forecast)
```

```python
m.plot(forecast, xlabel='Birth', ylabel='Total Win')
plt.savefig('klpga1.jpg')
```

```python
m.plot_components(forecast) #차트는 데이터프레임이 있어야 한다
plt.savefig('klpga2.jpg')
```
